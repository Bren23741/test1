## Section 2 — Classes & Objects (10 min)

### 2.1 Defining a class

```python
class Student:
    """Blueprint for a university student."""

    def __init__(self, name, gpa, year):
        # __init__ is the constructor: runs when you create an object
        # 'self' refers to this specific object
        self.name = name
        self.gpa  = gpa
        self.year = year

    def introduce(self):
        print(f"Hi, I'm {self.name}, year {self.year}, GPA {self.gpa}")

    def is_honor_roll(self):
        return self.gpa >= 3.5


# Creating objects (instances)
alice = Student('Alice', 3.8, 2)
bob   = Student('Bob',   3.2, 3)

alice.introduce()
bob.introduce()
print(f'Alice on honor roll? {alice.is_honor_roll()}')
print(f'Bob on honor roll?   {bob.is_honor_roll()}')
```

## Section 3 — Encapsulation (15 min)

**Encapsulation** = controlling access to an object's internal data.
Some data should only be changed through validated methods — not directly.

### The problem without it:

```python
alice = Student('Alice', 3.8, 2)
alice.gpa  = 99.9   # GPA can't be 99.9!
alice.year = -5     # Negative year makes no sense!
print(alice.gpa, alice.year)  # Python allows this — no protection
```

### 3.1 Python's encapsulation conventions

| Prefix | Meaning | Example |
|--------|---------|------|
| `name` | Public — anyone can access | `self.name` |
| `_name` | Protected — convention: internal use | `self._gpa` |
| `__name` | Private — name-mangled by Python | `self.__balance` |

```python
class BankAccount:

    def __init__(self, owner, initial_balance):
        self.owner     = owner              # public
        self.__balance = initial_balance    # private — double underscore

    def get_balance(self):
        return self.__balance

    def deposit(self, amount):
        if amount <= 0:
            raise ValueError('Deposit must be positive')
        self.__balance += amount
        print(f'Deposited ${amount}. Balance: ${self.__balance}')

    def withdraw(self, amount):
        if amount <= 0:
            raise ValueError('Withdrawal must be positive')
        if amount > self.__balance:
            raise ValueError('Insufficient funds')
        self.__balance -= amount
        print(f'Withdrew ${amount}. Balance: ${self.__balance}')


acc = BankAccount('Alice', 1000)
acc.deposit(500)
acc.withdraw(200)
print(f'Balance: ${acc.get_balance()}')
```

```python
# Try accessing __balance directly
try:
    print(acc.__balance)         # AttributeError!
except AttributeError as e:
    print(f'AttributeError: {e}')

# Python name-mangles it — accessible as _BankAccount__balance (but avoid this!)
print('Mangled access:', acc._BankAccount__balance)
```

```python
class Student:
    def __init__(self, name, gpa):
        self.name = name
        self._gpa  = gpa      # triggers the setter below

    def get_gpa(self):
        """Getter"""
        return self._gpa

    def set_gpa(self, value):
        """Setter: called when you write student.gpa = x"""
        if not (0.0 <= value <= 4.0):
            raise ValueError(f'GPA must be 0.0–4.0, got {value}')
        self._gpa = value


s = Student('Alice', 3.8)
print(s.get_gpa())      # getter — reads like a plain attribute

s.set_gpa(3.9)       # setter — validation runs
print(s.get_gpa())

try:
    s.set_gpa(99.9)  # rejected!
except ValueError as e:
    print(f'ValueError: {e}')
```

### 3.2 The Pythonic way: `@property`

Instead of writing `get_x()` / `set_x()` methods, Python has a cleaner approach:

```python
class Student:
    def __init__(self, name, gpa):
        self.name = name
        self.gpa  = gpa      # triggers the setter below

    @property
    def gpa(self):
        """Getter: called when you read  student.gpa"""
        return self._gpa

    @gpa.setter
    def gpa(self, value):
        """Setter: called when you write student.gpa = x"""
        if not (0.0 <= value <= 4.0):
            raise ValueError(f'GPA must be 0.0–4.0, got {value}')
        self._gpa = value


s = Student('Alice', 3.8)
print(s.gpa)      # getter — reads like a plain attribute

s.gpa = 3.9       # setter — validation runs
print(s.gpa)

try:
    s.gpa = 99.9  # rejected!
except ValueError as e:
    print(f'ValueError: {e}')
```

### Exercise 3.1
Create a `Temperature` class storing temperature in Celsius (`_celsius`).
- `celsius` property: getter + setter (validate: above -273.15)
- `fahrenheit` read-only property: converts and returns (formula: `C * 9/5 + 32`)

```python
class Temperature:
    def __init__(self, celsius):
        self.celsius = celsius   # triggers setter

    @property
    def celsius(self):
        return self._celsius  # getter

    @celsius.setter
    def celsius(self, value):
        if not (value > -273.15):  # setter with validation (> -273.15)
          raise ValueError(f'Must be over -273.15')
        self._celsius = value

    @property
    def fahrenheit(self):
        return self._celsius * 9/5 +32  # read-only conversion
        self._celsius = value



t = Temperature(100)
print(t.celsius)     # 100
print(t.fahrenheit)  # 212.0
t.celsius = -300     # ValueError
```

## Section 4 — Inheritance (15 min)

**Inheritance** = a child class automatically gets all attributes and methods from a parent class.

Models "is-a" relationships:
- A `GradStudent` **is a** `Student`
- A `Dog` **is an** `Animal`
- An `ElectricCar` **is a** `Car`

### 4.1 Basic inheritance

```python
class Animal:
    def __init__(self, name, species):
        self.name    = name
        self.species = species

    def eat(self):
        print(f'{self.name} is eating.')


    def speak(self):
        print(f'{self.name} makes a sound.')   # generic — will be overridden


class Dog(Animal):                             # Dog inherits from Animal
    def __init__(self, name, breed):
        super().__init__(name, species='Canis lupus')  # call parent __init__
        self.breed = breed

    def speak(self):                           # override parent method
        print(f'{self.name} says: Woof!')

    def fetch(self):                           # new method only in Dog
        print(f'{self.name} fetches the ball!')


class Cat(Animal):
    def __init__(self, name, indoor):
        super().__init__(name, species='Felis catus')
        self.indoor = indoor

    def speak(self):
        print(f'{self.name} says: Meow!')


dog = Dog('Rex', 'Labrador')
cat = Cat('Whiskers', indoor=True)

dog.eat()    # inherited from Animal
dog.speak()  # overridden in Dog
dog.fetch()  # Dog-specific
cat.speak()  # overridden in Cat
```

```python
from abc import ABC, abstractmethod   # FOR THE ABSTRACT FUNCTION IMPLEMENTATION

class Animal(ABC):
    def __init__(self, name, species):
        self.name    = name
        self.species = species

    def eat(self):
        print(f'{self.name} is eating.')

    @abstractmethod                   # ABSTRACT FUNCTION TO FORCE THE CHILD CLASS TO IMPLEMENT IT
    def speak(self):
        pass


class Cow(Animal):
    def __init__(self, name, breed, grassfed=True):
        super().__init__(name, breed)
        self.grass = grassfed


cow = Cow("Whitey", "Highlands")
```

### 4.2 Polymorphism — same interface, different behavior

```python
animals = [
    Dog('Rex',     'Labrador'),
    Cat('Whiskers', indoor=True),
    Dog('Buddy',   'Poodle')
]

# Each speaks in its own way — this is polymorphism
for animal in animals:
    animal.speak()
```

### 4.3 `isinstance` and `issubclass`

```python
dog = Dog('Rex', 'Labrador')

print(isinstance(dog, Dog))     # True
print(isinstance(dog, Animal))  # True  — Dog IS an Animal (inheritance!)
print(isinstance(dog, Cat))     # False
print(issubclass(Dog, Animal))  # True
print(issubclass(Cat, Dog))     # False
```

### 4.4 Multi-level inheritance + `super()`

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

    def greet(self):
        print(f"Hi, I'm {self.name}.")


class Student(Person):
    def __init__(self, name, age, student_id):
        super().__init__(name, age)
        self.student_id = student_id
        self._grades    = []

    def add_grade(self, grade):
        if 0 <= grade <= 100:
            self._grades.append(grade)

    def gpa(self):
        return round(sum(self._grades) / len(self._grades) / 25, 2) if self._grades else 0.0

    def greet(self):
        super().greet()    # call Person's greet first
        print(f'My student ID is {self.student_id}.')


class GradStudent(Student):
    def __init__(self, name, age, student_id, thesis_topic):
        super().__init__(name, age, student_id)
        self.thesis_topic = thesis_topic

    def greet(self):
        super().greet()
        print(f'My thesis: {self.thesis_topic}')


grad = GradStudent('Alice', 26, 'G1001', 'Neural Architecture Search')
grad.add_grade(92)
grad.add_grade(88)
grad.greet()
print(f'GPA: {grad.gpa()}')
```

## Section 5 — Mini Project: University System (15 min)

Combine **encapsulation** and **inheritance** in one system.

- `Person` — base class, validates email
- `Student(Person)` — private grades, GPA, transcript
- `Professor(Person)` — department, courses taught
- `TAStudent(Student)` — is both a student and a TA

```python
class Person:
    def __init__(self, name, email):
        self.name  = name
        self.email = email    # triggers property setter

    @property
    def email(self):
        return self._email

    @email.setter
    def email(self, value):
        if '@' not in value:
            raise ValueError(f'Invalid email: {value}')
        self._email = value

    def __repr__(self):
        return f'{self.__class__.__name__}({self.name!r})'


class Student(Person):
    def __init__(self, name, email, student_id):
        super().__init__(name, email)
        self.student_id = student_id
        self.__grades   = []   # private

    def add_grade(self, course, score):
        if not (0 <= score <= 100):
            raise ValueError('Score must be 0-100')
        self.__grades.append({'course': course, 'score': score})

    def gpa(self):
        if not self.__grades:
            return 0.0
        avg = sum(g['score'] for g in self.__grades) / len(self.__grades)
        return round(avg / 25, 2)   # 100 -> 4.0

    def transcript(self):
        print(f'\n=== Transcript: {self.name} ({self.student_id}) ===')
        for g in self.__grades:
            print(f"  {g['course']:<20} {g['score']}")
        print(f"  {'GPA':<20} {self.gpa()}")


class Professor(Person):
    def __init__(self, name, email, department):
        super().__init__(name, email)
        self.department      = department
        self._courses_taught = []

    def assign_course(self, course_name):
        self._courses_taught.append(course_name)
        print(f'{self.name} assigned to: {course_name}')

    def profile(self):
        print(f'\nProf. {self.name} | {self.department}')
        print(f"  Courses: {', '.join(self._courses_taught) or 'None'}")


class TAStudent(Student):
    """Graduate student who is also a teaching assistant."""

    def __init__(self, name, email, student_id, supervising_professor):
        super().__init__(name, email, student_id)
        self.supervising_professor = supervising_professor

    def profile(self):
        print(f'\nTA: {self.name} | Supervised by: {self.supervising_professor.name}')
        self.transcript()


# --- Run the system ---
prof = Professor('Dr. ElSaid', 'elsaida@uncw.edu', 'Computer Science')
prof.assign_course('Machine Learning')
prof.assign_course('AI Systems')
prof.profile()

alice = Student('Alice', 'alice@uncw.edu', 'S1001')
alice.add_grade('Machine Learning', 92)
alice.add_grade('Data Structures',  85)
alice.add_grade('Calculus II',      78)
alice.transcript()

ta = TAStudent('Bob', 'bob@uncw.edu', 'G2001', supervising_professor=prof)
ta.add_grade('Research Methods', 95)
ta.add_grade('AI Systems',       90)
ta.profile()
```

```python
class Student:
    def __init__(self, student_id):
        self.student_id = student_id

class Teacher:
    def __init__(self, department):
        self.department = department

class TAStudent(Student, Teacher):
    def __init__(self, name, student_id, department):
        Student.__init__(self, student_id)    # call each parent directly
        Teacher.__init__(self, department)    # by name, not super()
        self.name = name


ta = TAStudent("Alice", "S001", "CS")
print(ta.student_id)   # S001
print(ta.department)   # CS
```

```python
class Person:
    def __init__(self, name, **kwargs):
        self.name = name
        super().__init__(**kwargs)      # passes remaining kwargs up the chain

class Student(Person):
    def __init__(self, student_id, **kwargs):
        self.student_id = student_id
        super().__init__(**kwargs)      # passes remaining kwargs up

class Teacher(Person):
    def __init__(self, department, **kwargs):
        self.department = department
        super().__init__(**kwargs)      # passes remaining kwargs up

class TAStudent(Student, Teacher):
    def __init__(self, name, student_id, department):
        super().__init__(name=name, student_id=student_id, department=department)


ta = TAStudent("Alice", "S001", "CS")
print(ta.name)        # Alice
print(ta.student_id)  # S001
print(ta.department)  # CS
```

## Summary

| Concept | Key idea | Python syntax |
|---------|----------|---------------|
| **Class** | Blueprint for objects | `class MyClass:` |
| **Object** | Instance of a class | `obj = MyClass()` |
| **`__init__`** | Constructor | `def __init__(self, ...)` |
| **Encapsulation** | Protect & validate data | `_x`, `__x`, `@property` |
| **Inheritance** | Child reuses parent | `class Child(Parent):` |
| **`super()`** | Call parent method | `super().__init__(...)` |
| **Override** | Redefine parent method | Same method name in child |
| **Polymorphism** | Same call, different behavior | Emerges from overriding |
